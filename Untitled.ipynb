```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time 
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense,MaxPool2D, Dropout, Flatten, Conv2D, GlobalAveragePooling2D, Activation  
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from random import choice, shuffle
from scipy import stats as st
from collections import deque
```

```python
def gather_data(num_samples):
  global rock, paper, scissor, nothing
  cap=cv2.VideoCapture(0,cv2.CAP_DSHOW)
  tr=False
  counter=0
  boxShape=234
  width=int(cap.get(3))
  while True:
    r, frame_BoundinfBox= cap.read()
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    frame_BoundinfBox=cv2.flip(frame_BoundinfBox,1)
    if not r:
      break
    if counter==num_samples:
      tr=not tr
      counter=0
    cv2.rectangle(frame_BoundinfBox,(width-boxShape, 0),(width, boxShape),(0,250,150),2)
    cv2.namedWindow("Collecting Images", cv2.WINDOW_NORMAL)
    if tr:
      R= frame_BoundinfBox[5: boxShape-5, width-boxShape+5:width-5]
      eval(class_name).append([R, class_name])
      counter+=1
      displayMSG="Gathers images{}:{}".format(class_name, counter)
    else:
      displayMSG= "Press 'r' to collect rock sample, 'p' for paper, 's' for scissor and 'n' for nothing"
    cv2.putText(frame_BoundinfBox,displayMSG,(3,350), cv2.FONT_HERSHEY_SIMPLEX,0.45,(0,0,255),1, cv2.LINE_AA )
    cv2.imshow("Collecting images", frame_BoundinfBox)

    k = cv2.waitKey(1)

 
    if k==ord('r'):
      tr= not tr
      class_name='rock'
      rock=[]
    if k==ord('p'):
      tr= not tr
      class_name='paper'
      paper=[]
    if k==ord('s'):
      tr= not tr
      class_name='scissor'
      scissor=[]
    if k==ord('n'):
      tr= not tr
      class_name='nothing'
      nothing=[]
    if k == ord('q'):
      break
           
  cap.release()
  cv2.destroyAllWindows()
n = 100
gather_data(n)
```

```python
LB = [tupl[1] for tupl in rock] + [tupl[1] for tupl in paper] + [tupl[1] for tupl in scissor] +[tupl[1] for tupl in nothing]
images = [tupl[0] for tupl in rock] + [tupl[0] for tupl in paper] + [tupl[0] for tupl in scissor] +[tupl[0] for tupl in nothing]
images = np.array(images, dtype="float") / 255.0
print(' About {} Total images , {} Total Labels'.format(len(LB), len(images)))
encoding_datas = LabelEncoder()
Int_LB = encoding_datas.fit_transform(LB)
one_hot_LB = to_categorical(Int_LB, 4)
(trainX, testX, trainY, testY) = train_test_split(images, one_hot_LB, test_size=0.25, random_state=50)
images = []
```

```python
Shape_of_Image = 224
NasNetting = tf.keras.applications.NASNetMobile( input_shape=(Shape_of_Image, Shape_of_Image, 3), include_top=False, weights='imagenet')
NasNetting.trainable = False
x = NasNetting.output
x = GlobalAveragePooling2D()(x)
x = Dense(712, activation='relu')(x) 
x = Dropout(0.40)(x)
preds = Dense(4,activation='softmax')(x) 
model = Model(inputs=NasNetting.input, outputs=preds)
print ("Total {} Layers in the Model".format(len(model.layers[:])))
```

```python
augment = ImageDataGenerator( 
    
        rotation_range=30,
        zoom_range=0.25,
        width_shift_range=0.10,
        height_shift_range=0.10,
        shear_range=0.10,
        horizontal_flip=False,
        fill_mode="nearest"
)
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
epochs = 15
BSize = 20
history = model.fit(x=augment.flow(trainX, trainY, batch_size=BSize), validation_data=(testX, testY), 
steps_per_epoch= len(trainX) // BSize, epochs=epochs)
```

```python
accuracyOfModel = history.history['accuracy']
val_accuracyOfModel = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracyOfModel))
plt.plot(epochs, accuracyOfModel, 'b', label='Training accuracy Of Model')
plt.plot(epochs, val_accuracyOfModel, 'r', label='Validation accuracy Of Model')
plt.title('Training accuracy Of Model')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training loss')
plt.legend()
plt.show()
```

```python
def findout_winner(user_move, Computer_move):    
    if user_move == Computer_move:
        return "Tie"    
    
    elif user_move == "rock" and Computer_move == "scissor":
        return "User"
    
    elif user_move == "rock" and Computer_move == "paper":
        return "Computer"
    
    elif user_move == "scissor" and Computer_move == "rock":
        return "Computer"
    
    elif user_move == "scissor" and Computer_move == "paper":
        return "User"
    
    elif user_move == "paper" and Computer_move == "rock":
        return "User"
    
    elif user_move == "paper" and Computer_move == "scissor":
        return "Computer"
```

```python
def show_winner(user_socre, computer_score):    
    
    if user_score > computer_score:
        img = cv2.imread("images/youwin.jpg")
        
    elif user_score < computer_score:
        img = cv2.imread("images/comwins.jpg")
        
    else:
        img = cv2.imread("images/draw.jpg")
        
    cv2.putText(img, "Press 'ENTER' to play again, else exit",
                (150, 530), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 3, cv2.LINE_AA)
    
    cv2.imshow("Rock Paper Scissors", img)
    
    # If enter is pressed.
    k = cv2.waitKey(0)
    
    # If the user presses 'ENTER' key then return TRUE, otherwise FALSE
    if k == 13:
       return True

    else:
        return False
```

```python
def display_computer_move(computer_move_name, frame):
    
    icon = cv2.imread( "images/{}.png".format(computer_move_name), 1)
    print(icon)
    try:
        print("Ekhane ami vai")
        icon = cv2.resize(icon, (224,224), )
        print(icon.shape)
        print(icon)
    except:
        print("Kire vai eto jhamela ken!!!")

    R = frame[0:224, 0:224]
 
    mask = icon[:,:,-1] 
    mask = cv2.threshold(mask, 1, 255, cv2.THRESH_BINARY)[1]
    icon_bgr = icon[:,:,:3]     
    img1_bg = cv2.bitwise_and(R, R, mask = cv2.bitwise_not(mask))
    img2_fg = cv2.bitwise_and(icon_bgr, icon_bgr, mask = mask)
    combined = cv2.add(img1_bg, img2_fg)
    frame[0:224, 0:224] = combined
    return frame
```

```python
cap = cv2.VideoCapture(0)
box_size = 234
width = int(cap.get(3))
attempts = 5
computer_move_name= "nothing"
final_user_move = "nothing"

operations = ['nothing', 'paper', 'rock', 'scissor']

computer_score, user_score = 0, 0
rect_color = (255, 0, 0)
hand_inside = False
total_attempts = attempts
confidence_threshold = 0.70

smooth_factor = 5
de = deque(['nothing'] * 5, maxlen=smooth_factor)

while True:
    
    ret, frame = cap.read()
    
    if not ret:
        break
        
    frame = cv2.flip(frame, 1)
           
    cv2.namedWindow("Rock Paper Scissors", cv2.WINDOW_NORMAL)
    R = frame[5: box_size-5 , width-box_size + 5: width -5]
    R = np.array([R]).astype('float64') / 255.0
    pred = model.predict(R)
    move_code = np.argmax(pred[0])
    user_move = operations[move_code]
    prob = np.max(pred[0])
    if prob >= confidence_threshold:
        de.appendleft(user_move)
        try:
            final_user_move = st.mode(de)[0][0] 
            
        except StatisticsError:
            print('Stats error')
            continue        
        if final_user_move != "nothing" and hand_inside == False:
            hand_inside = True 
            computer_move_name = choice(['rock', 'paper', 'scissor'])
            winner = findout_winner(final_user_move, computer_move_name)
            display_computer_move(computer_move_name, frame)

            total_attempts -= 1
            if winner == "Computer":
                computer_score +=1
                rect_color = (0, 0, 255)

            elif winner == "User":
                user_score += 1;
                rect_color = (0, 250, 0)                
            
            elif winner == "Tie":
                rect_color = (255, 250, 255)     
            if total_attempts == 0:
                
                play_again = show_winner(user_score, computer_score)
                if play_again:
                    user_score, computer_score, total_attempts = 0, 0, attempts
                else:
                    break
        
        elif final_user_move != "nothing" and hand_inside == True:
            display_computer_move(computer_move_name, frame)

        elif final_user_move == 'nothing':            
            hand_inside = False
            rect_color = (255, 0, 0) 

    cv2.putText(frame, "Your Move: " + final_user_move,
                    (420, 270), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)

    cv2.putText(frame, "Computer's Move: " + computer_move_name,
                (2, 270), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)

    cv2.putText(frame, "Your Score: " + str(user_score),
                    (420, 300), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)
    cv2.putText(frame, "Computer Score: " + str(computer_score),
                    (2, 300), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)
    
    cv2.putText(frame, "Attempts left: {}".format(total_attempts), (190, 400), cv2.FONT_HERSHEY_COMPLEX, 0.7,
                (100, 2, 255), 1, cv2.LINE_AA)    
    
    cv2.rectangle(frame, (width - box_size, 0), (width, box_size), rect_color, 2) 
    cv2.imshow("Rock Paper Scissors", frame)
    k = cv2.waitKey(10)
    if k == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
